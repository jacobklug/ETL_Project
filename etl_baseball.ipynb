```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
standings_file = "Standings2018.csv"
standings_df = pd.read_csv(standings_file)
standings_df = standings_df[["Tm", "W", "L", "Rdiff", "Home", "Road"]]
standings_df = standings_df.rename(columns = {"Tm": "Team"})
standings_df.head()
```

```python
attendance_file = "attendance2018.csv"
attendance_df = pd.read_csv(attendance_file)
attendance_df = attendance_df[["Tm", "Attendance", "Attend/G", "Est. Payroll"]]
attendance_df = attendance_df.rename(columns = {"Tm": "Team"})
attendance_df.head()
```

```python
valuations_file = "valuations.csv"
valuations_df = pd.read_csv(valuations_file)
valuations_df = valuations_df[["Team", "Current Value", "Revenue"]]
valuations_df = valuations_df.replace(["FLA","TB", "SD", "WAS", "KAN"], ["MIA", "TBR", "SDP", "WSN", "KCR"])
valuations_df.head()
```

```python
baseball_merged = pd.merge(standings_df, attendance_df, how = "left", on = ["Team"])
baseball_merged.head()
```

```python
baseball_merged = pd.merge(baseball_merged, valuations_df, how = "left", on = ["Team"])
baseball_merged.head()
```

```python
engine = create_engine('postgres://postgres:postgres@localhost:5432/etl_baseball')
connection = engine.connect()
baseball_merged.to_sql('baseball_merged', engine)
```
